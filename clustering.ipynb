```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import process_data as proc
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler, StandardScaler

```

```python
# Load dataset
df = pd.read_csv('total_df.csv')

# Index dataframe with LNR id
X = df.set_index('LNR').astype(np.int8)

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)

# KMeans Clustering
kmeans = KMeans(n_clusters=6, random_state=42, max_iter=700)

#print("Clustering ... \n")
kmeans.fit(X_scaled)
clusters = kmeans.predict(X_scaled)
X['cluster'] = clusters

# Store DataFrame
print("Storing clustered data...")
X.to_csv('clustered_df.csv', header=X.columns)
```

```python
# Percentage of customer per cluster
customer_id=np.loadtxt('customers_id.out', int)

# X = ps.read_csv('clustered_df.csv')
X_customers = X.loc[np.intersect1d(customer_id, X.index)]
custom_prct=X_customers['cluster'].value_counts() / X_customers['cluster'].value_counts().sum()

general_prct=X['cluster'].value_counts() / X['cluster'].value_counts().sum()

```

```python
# Difference between customers and population

custom_prct = custom_prct.sort_index()*100
custom_prct.name = 'Customer'

general_prct = general_prct.sort_index()*100
general_prct.name = 'Population'

diffs = (custom_prct - general_prct)
diffs.name = 'Difference'

x = pd.merge(general_prct, custom_prct, left_index=True, right_index=True)
xdf = pd.merge(x, diffs, left_index=True, right_index=True)
xdf.index.name = 'Cluster'

xdf.style.bar(subset= diffs.name, align='mid', color=['#d65f5f', '#5fba7d'])
```
